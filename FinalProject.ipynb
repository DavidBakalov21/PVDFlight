```python
import pandas as pd
```

```python
df=pd.read_csv('Data_Train.csv')
df
```

```python
df['Total_Stops'].unique()
```

```python
df.dropna(subset=['Total_Stops'], inplace=True)
df['Total_Stops'].unique()
```

```python
df['Total_Stops']=df['Total_Stops'].apply(lambda x: '0' if x == 'non-stop' else x)
df['Total_Stops']=pd.to_numeric(df['Total_Stops'].apply(lambda x: x.split(' ')[0]))
df['Total_Stops'].unique()
```

```python
df.dropna(subset=['Price'], inplace=True)
df.dropna(subset=['Duration'], inplace=True)
df.dropna(subset=['Arrival_Time'], inplace=True)
df.dropna(subset=['Route'], inplace=True)
df.dropna(subset=['Dep_Time'], inplace=True)
df.dropna(subset=['Destination'], inplace=True)
df.dropna(subset=['Source'], inplace=True)
df.dropna(subset=['Date_of_Journey'], inplace=True)
df.dropna(subset=['Airline'], inplace=True)
df['Duration'].unique()
```

```python
def duration_to_minutes(duration):
    if 'h' in duration:
        hours_part = duration.split('h')[0]
        minutes_part = duration.split('h')[1]
        hours = int(hours_part)
        if minutes_part!='':
            minutes = int(minutes_part.split('m')[0])
        else:
            minutes=0
    else:
        hours=0
        minutes=int(duration.split('m')[0])
    total_minutes = hours * 60 + minutes

    return total_minutes

df['Duration'] = df['Duration'].apply(duration_to_minutes)
df['Duration'].unique()
```

```python
def only_month(date):
    data=date.split(' ')
    if len(data)==3:
        return data[2]
    else:
        return 'Unknown'


df['Arrival_Time']=df['Arrival_Time'].apply(only_month)
df['Arrival_Time'].unique()
```

```python
def only_day_num(date):
    data=date.split('/')
    if len(data)==3:
        day=data[0]
        if day[0]=='0':
            return day[1]
        return day
    else:
        return 'Unknown'

df['Day_of_Journey']=df['Date_of_Journey'].apply(only_day_num)
df['Day_of_Journey'].unique()
```

```python
df
```

```python
def num_to_name(num):
    if '5' in num or num=='May':
        return 'May'
    if '4' in num or num=='Apr':
        return 'Apr'
    if '3' in num or num=='Mar':
        return 'Mar'
    if '6' in num or num=='Jun':
        return 'Jun'
        

mask = (df['Arrival_Time'] == 'Unknown')
split_values = df['Date_of_Journey'].str.split('/')
df.loc[mask, 'Arrival_Time'] = split_values.apply(lambda x: x[1])
df['Arrival_Time']=df['Arrival_Time'].apply(num_to_name)
df
```

```python
df.rename(columns={'Arrival_Time': 'Arrival_Time(month)'}, inplace=True)
df.rename(columns={'Duration': 'Duration(minutes)'}, inplace=True)
df
```

```python
df.to_csv('FlightCleaned.csv', index=False)
```
